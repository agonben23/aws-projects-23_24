```python
import pandas as pd
```

```python
char_stats_df = pd.read_csv("./characters_stats.csv").drop(labels=["Alignment","Total"],axis=1)
```

```python
char_stats_df
```

```python
super_power_df = pd.read_csv("./superheroes_power_matrix.csv")
```

```python
super_power_df
```

```python
marvel_characters = pd.read_csv("./marvel_characters_info.csv").drop(labels=["ID","Publisher"],axis=1)
marvel_characters.replace('-', pd.NA, inplace=True)
marvel_characters = marvel_characters.dropna()
marvel_characters = marvel_characters.reset_index(drop=True)

mask = marvel_characters['Height'] == -99.0
marvel_characters.loc[mask, 'Height'] = pd.NA

mask = marvel_characters['Weight'] == -99.0
marvel_characters.loc[mask, 'Weight'] = pd.NA

df_no_nuls = marvel_characters.dropna()

mean_heights = df_no_nuls.groupby(['Gender', 'Race'])['Height'].mean()


def replace_with_mean_height(row):
    if pd.isna(row['Height']):
        return mean_heights.get(row['Race'],pd.NA)
    else:
        return row['Height']


marvel_characters['Height'] = marvel_characters.apply(replace_with_mean_height, axis=1)


mean_weights = df_no_nuls.groupby(['Race'])['Weight'].mean()

def replace_with_mean_weight(row):
    if pd.isna(row['Weight']):
        return mean_weights.get(row['Race'],pd.NA)
    else:
        return row['Weight']


marvel_characters['Weight'] = marvel_characters.apply(replace_with_mean_weight, axis=1)

marvel_characters = marvel_characters.dropna()
marvel_characters = marvel_characters.reset_index(drop=True)
```

```python
marvel_characters
```

```python
final_df = marvel_characters.merge(super_power_df,on="Name",how="inner")
```

```python
final_df
```

```python
# Convertir las columnas de categoría a tipo categoría
columns_to_convert = ["Gender", "EyeColor", "Race", "HairColor", "SkinColor"]
for column in columns_to_convert:
    final_df[column] = final_df[column].astype('category')

# Mapear las categorías a valores numéricos
for column in columns_to_convert:
    final_df[column] = final_df[column].cat.codes
```

```python
final_df
```

```python
# Algoritmos
```

```python
from sklearn.cluster import KMeans

names = final_df[["Name"]]
X = final_df.drop(labels=["Name","Alignment"],axis=1)
Y = final_df[["Alignment"]]

kmeans = KMeans(n_clusters=3,n_init=10)
kmeans.fit(X)

labels = kmeans.labels_

centroids = kmeans.cluster_centers_
```

```python
labels
```

```python
from sklearn.metrics import silhouette_score
from matplotlib import pyplot as plt

np_X = X.values if isinstance(X, pd.DataFrame) else X

score = silhouette_score(np_X, labels)
print("Silhouette score:", score)
plt.figure()

plt.scatter(np_X[:,0], np_X[:,1], c=labels)
plt.title("Clustering results")
plt.show()
```
